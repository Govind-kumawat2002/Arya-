```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data =sns.load_dataset("tips") # load dataset from seaborn
```

```python
data.head()
```

```python
plt.plot(data['total_bill'])
plt.show()
```

```python
sns.lineplot(data['total_bill']) # line plot of total_bill
```

```python
sns.scatterplot(x = 'total_bill' ,y = 'tip',data=data) # we are doing scatter plot of total_bill and tip
```

```python
sns.histplot(data['total_bill']) # histogram of total_bill
```

```python
plt.hist(data['total_bill']) # histogram of total_bill
```

```python
sns.kdeplot(data['total_bill']) # kde plot of total_bill
```

```python
data
```

```python
# catplot 
# relplot # scatter plot
```

```python
sns.relplot(x = 'total_bill' ,y = 'tip',data=data) # we are doing scatter plot of total_bill and tip
```

```python
data
```

```python
data['time'].unique()
```

```python
sns.relplot(x ='total_bill', y = 'tip', hue = 'time', data = data)
```

```python
sns.relplot(x = 'total_bill' ,y = 'tip',hue = 'time',style='smoker',data=data) # we are doing scatter plot of total_bill and tip
```

```python
sns.relplot(x = 'total_bill' ,y = 'tip',hue = 'time',style='smoker',data=data,kind='scatter',col='smoker') # we are doing scatter plot of total_bill and tip
```

```python
sns.relplot(x='total_bill', y = 'tip', hue = 'time', data = data,kind='scatter',style='smoker',row='time')
```

```python
data['day'].unique()
```

```python
sns.relplot(x='total_bill', y = 'tip', hue = 'time', data = data,kind='scatter',col='day')
```

```python
sns.relplot(x='total_bill', y = 'tip',data=data)
plt.title("total bill vs tip")
plt.xlabel("total bill",fontsize=20,color='red')
plt.ylabel("tip",fontsize=20,color='red')
plt.show()
```
